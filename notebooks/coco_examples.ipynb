```python
from pycocotools.coco import COCO
import numpy as np
```

```python
coco_data = COCO("../input/data/train.json")
```

```python
image_id = coco_data.getImgIds(imgIds=0)
image_id
```

```python
image_infos = coco_data.loadImgs(image_id)[0]
image_infos
```

```python
ann_ids = coco_data.getAnnIds(imgIds=image_infos["id"])
ann_ids
```

```python
anns = coco_data.loadAnns(ann_ids)
anns
```

```python
cat_ids = coco_data.getCatIds()
cat_ids
```

```python
cats = coco_data.loadCats(cat_ids)
cats
```
